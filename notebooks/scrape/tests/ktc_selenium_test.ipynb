```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Firefox(executable_path=r'C:/Users/jrior/.ipython/drivers/geckodriver.exe')
driver.get("https://keeptradecut.com/dynasty-rankings/players/jonathan-taylor-634")
#assert "Python" in driver.title
alltime = driver.find_element(By.ID, "all-time")
alltime.click()
content = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(content)
valgraphsoup = soup.find("div", {"id":"pd-value-graph"})
```

```python
dates = []
vals = []
hovergroups = valgraphsoup.find_all("g", {"class":"hoverGroup"})
for hg in hovergroups:
    dates.append(hg.find("text", {"class":"hoverDate"}).text)
    vals.append(hg.find("text", {"class":"graphVal hoverVal"}).text)

pairs = list(zip(dates, vals))
list(pairs)
```

```python
hovergroups = valgraphsoup.find_all("g", {"class":"hoverGroup"})
pd.DataFrame({'name':['Matthew Stafford' for hg in hovergroups],
             'date':[hg.find("text", {"class":"hoverDate"}).text for hg in hovergroups],
             'value':[hg.find("text", {"class":"graphVal hoverVal"}).text for hg in hovergroups]})
```
